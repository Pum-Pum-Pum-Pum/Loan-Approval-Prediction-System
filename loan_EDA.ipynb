## 1) Imports and setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 100)
sns.set_style('whitegrid')
RANDOM_STATE = 42
```

## 2) Load data

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

print('Train shape:', train.shape)
print('Test shape :', test.shape)
train.head()
```

## 3) Basic schema check
Goal: verify target column and train/test column alignment.

```python
print('Train columns:', list(train.columns))
print('Test columns :', list(test.columns))

target_col = 'Loan_Status'
feature_cols = [c for c in train.columns if c != target_col]

print('\nTarget column:', target_col)
print('Feature count:', len(feature_cols))
print('Test has same feature columns:', feature_cols == list(test.columns))
```

## 4) Data types + missing values
Goal: identify columns needing preprocessing in ML pipeline.

```python
dtype_df = pd.DataFrame({
    'column': train.columns,
    'dtype': [str(train[c].dtype) for c in train.columns]
})
dtype_df
```

```python
missing_df = pd.DataFrame({
    'column': train.columns,
    'missing_count': [train[c].isna().sum() for c in train.columns],
    'missing_pct': [round(train[c].isna().mean() * 100, 2) for c in train.columns]
}).sort_values('missing_pct', ascending=False)

missing_df
```

## 5) Target distribution (class imbalance check)
Goal: understand why accuracy alone can be misleading.

```python
target_counts = train[target_col].value_counts()
target_ratio = (train[target_col].value_counts(normalize=True) * 100).round(2)

print('Target counts:\n', target_counts)
print('\nTarget ratio (%):\n', target_ratio)
```

```python
plt.figure(figsize=(5,4))
sns.countplot(data=train, x=target_col, order=target_counts.index)
plt.title('Loan_Status Distribution')
plt.show()
```

## 6) Quick categorical sanity checks
Goal: detect special values needing cleaning (example: Dependents = '3+').

```python
cat_cols = train.select_dtypes(include='object').columns.tolist()
for col in cat_cols:
    print(f'\n{col} unique values:')
    print(sorted(train[col].dropna().unique().tolist()))
```
